# Bifurcation Diagrams

## Logistic

```python
from LogisticMapLCE import logistic_lce
rr = np.sort(np.random.uniform(3.5, 4, 1000))
logistic_dataset = [logistic_lce(nIterates=1000, includeTrajectory=True, mapParams=dict(r=r)) for r in rr]
```

```python
for logistic_data in logistic_dataset:
    trajectory = logistic_data["trajectory"]
    r = logistic_data["params"]["r"] * np.ones_like(trajectory)
    plt.scatter(r, trajectory, c="C0", s=0.0001)
plt.title("Logistic map bifurcation diagram $x_n$ vs $r$")
plt.xlabel("Control parameter $r$")
plt.ylabel("State variable $x_n$")
plt.savefig("./figures/logistic_bifurcation.png", dpi=300)
plt.show()
```

```python
for henon_data in henon_dataset[::4]:
    trajectory = henon_data["trajectory"][:,1]
    a = henon_data["params"]["a"] * np.ones_like(trajectory)
    plt.scatter(a, trajectory, c="C0", s=0.0005)
plt.title("Hnon map bifurcation diagram $y_n$ vs $a$: $b=0.3$")
plt.xlabel("Control parameter $a$")
plt.ylabel("State variable $y_n$")
plt.savefig("./figures/henon_bifurcation.png", dpi=300)
plt.show()
```

# Lyapunov Exponents over Bifurcation Diagrams

## Logistic

```python
from LogisticMapLCE import logistic_lce
rr = np.sort(np.random.uniform(3.5, 4, 1000))
logistic_dataset = [logistic_lce(nIterates=1000, includeTrajectory=True, mapParams=dict(r=r)) for r in rr]
```

```python
fig, ax1 = plt.subplots()
fig.suptitle("Logistic map $\lambda_{\max}$ vs $r$")

color1 = "gray"
for logistic_data in logistic_dataset:
    trajectory = logistic_data["trajectory"]
    r = logistic_data["params"]["r"] * np.ones_like(trajectory)
    ax1.scatter(r, trajectory, c=color1, s=0.0001, alpha=0.2)
ax1.tick_params(axis='y', labelcolor=color1)
ax1.set_xlabel("Control parameter $r$")
ax1.set_ylabel("State variable $x_n$", c=color1)

ax2 = ax1.twinx()

lyapunovs = [logistic_data["lce"] for logistic_data in logistic_dataset]
rr = [logistic_data["params"]["r"] for logistic_data in logistic_dataset]
color2 = "red"
ax2.plot(rr, lyapunovs, c=color2, lw=0.75)
ax2.axhline(c=color2, linestyle="-.", lw=0.5)
ax2.set_ylabel("Lyapunov exponent $\lambda_{\max}$", color=color2)
ax2.tick_params(axis="y", labelcolor=color2)
plt.savefig("./figures/logistic_bifurcation_with_lyapunovs.png", dpi=300)
plt.show()
```

## Data sets

```python
from numpy.random import MT19937
from numpy.random import RandomState
from numpy.random import SeedSequence
SEED = 42   
```

```python
from LogisticMapLCE import logistic_lce
randomState = RandomState(MT19937(SeedSequence(SEED)))
logistic_r_values = np.sort(randomState.uniform(3.4, 4, 1000))
logistic_dataset = [logistic_lce(nIterates=1000, includeTrajectory=True, mapParams=dict(r=r)) for r in logistic_r_values]
logistic_trajectories = np.array([logistic_data["trajectory"] for logistic_data in logistic_dataset])
logistic_lces = np.array([logistic_data["lce"] for logistic_data in logistic_dataset])
```

```python
np.sum(logistic_lces > 0)
```

```python

from HenonMapLCE import henon_lce
randomState = RandomState(MT19937(SeedSequence(SEED)))
henon_a_values = np.sort(randomState.uniform(0.8, 1.4, 1000))
henon_dataset = [henon_lce(nIterates=1000, includeTrajectory=True, mapParams=dict(a=a, b=0.3)) for a in henon_a_values]
henon_trajectories = np.array([henon_data["trajectory"][:,1] for henon_data in henon_dataset])
henon_lces = np.array([henon_data["lce"] for henon_data in henon_dataset])
```

```python
np.sum(henon_lces > 0)
```

```python

from LorenzODELCE import lorenz_lce
import ipyparallel as ipp
```

```python

clients = ipp.Client()
dview = clients.direct_view()
dview.push(dict(lorenz_lce=lorenz_lce))
```

```python

randomState = RandomState(MT19937(SeedSequence(SEED)))
lorenz_rho_values = np.sort(randomState.uniform(90, 105, 1000))

lorenz_dataset = dview.map_async(lambda r: lorenz_lce(nIterates=1000, includeTrajectory=True, lorenzParams=dict(sigma=10, b=8/3, R=r)), lorenz_rho_values)
```

```python

lorenz_trajectories = np.array([lorenz_data["trajectory"][:,0] for lorenz_data in lorenz_dataset])
lorenz_lces = np.array([lorenz_data["lce"] for lorenz_data in lorenz_dataset])
```

```python
np.sum(lorenz_lces > 0)
```

```python
plt.plot(lorenz_rho_values, lorenz_lces, lw=0.7)
plt.title("Lorenz map $\lambda_{\max}$")
plt.xlabel("Control parameter rho")
plt.ylabel("Lyapunov exponent $\lambda_{\max}$")
plt.show()
```

## Rosenstein and Eckmann estimates (Python nolds)

```python
from nolds import lyap_e, lyap_r
```

```python
lorenz_rosenstein_estimates = [lyap_r(ts) for ts in lorenz_trajectories]
```

```python
lorenz_eckmann_estimates = [lyap_e(ts)[0] for ts in lorenz_trajectories]
```

```python
from TimeSeriesMergeTree import TimeSeriesMergeTree as TSMT
```

```python
lorenz_tsmts = [TSMT(ts) for ts in lorenz_trajectories]
```

```python
henon_tsmts = [TSMT(ts) for ts in henon_trajectories]
```

```python
import ipyparallel as ipp
clients = ipp.Client()
dview = clients.load_balanced_view()
```

```python
lorenz_interleaving_estimates = dview.map_async(lambda tsmt: (tsmt[0], tsmt[1].interleaving_divergence), zip(lorenz_rho_values, lorenz_tsmts), ordered=False)
```

```python
henon_interleaving_estimates = dview.map_async(lambda tsmt: (tsmt[0], tsmt[1].interleaving_divergence), zip(henon_a_values[::5], henon_tsmts[::5]), ordered=False)
```

```python
clients.shutdown(hub=True)
```

```python
results = []
counter = 0
for estimate in henon_interleaving_estimates:
    results.append(estimate)
    print(len(results), estimate[0], estimate[1], henon_lces[counter][0])
    counter += 5
```

```python
data = np.array(results)
```

```python
data = data[data[:,0].argsort()]
```
