```python
import numpy as np
import matplotlib.pylab as plt
```

# Trajectories

```python
from LogisticMapLCE import logistic_lce
logistic_data = logistic_lce(nIterates=500, includeTrajectory=True)
logistic_data
```

```python
plt.plot(logistic_data["trajectory"], lw=0.5)
plt.xlabel("$n-100$")
plt.ylabel("$x_n$")
plt.title("Logistic map trajectory $x_n$: $r=4, x_0=0.2$")
plt.savefig("./figures/logistic_trajectory.pdf")
plt.show()
```

```python
from HenonMapLCE import henon_lce
henon_data = henon_lce(nIterates=500, includeTrajectory=True)
henon_data
```

```python
plt.plot(henon_data["trajectory"][:,1], lw=0.5)
plt.xlabel("$n-100$")
plt.ylabel("$y_n$")
plt.title("Hénon map trajectory $y_n$: $(a, b) = (1.4, 0.3), (x_0, y_0) = (0.1, 0.3)$")
plt.savefig("./figures/henon_trajectory.pdf")
plt.show()
```

```python
plt.scatter(henon_data["trajectory"][:,0], henon_data["trajectory"][:,1], s=1)
plt.xlabel("$x_n$")
plt.ylabel("$y_n$")
plt.title("Hénon map states $(x_n,y_n), 100\leq n\leq 500$:\n $(a, b) = (1.4, 0.3), (x_0, y_0) = (0.1, 0.3)$")
plt.savefig("./figures/henon_attractor.pdf")
plt.show()
```

```python
from LorenzODELCE import lorenz_lce
lorenz_data = lorenz_lce(nIterates=5000, includeTrajectory=True)
lorenz_data
```

```python
dt = 0.01
nIterates = lorenz_data["iterates"]["trajectory"]["nIterates"]
nTransients = lorenz_data["iterates"]["trajectory"]["nTransients"]
t0 = nTransients*dt
t1 = nIterates*dt + t0
times = np.linspace(t0, t1, nIterates+1)
times
```

```python
plt.plot(times, lorenz_data["trajectory"][:,0], lw=0.8)
plt.xlabel("$t$")
plt.ylabel("$x(t)$")
plt.title("Lorenz system trajectory $x(t), 1\leq t\leq 51$:\n $(σ, β, ρ) = (10, 8/3, 28), (x_0, y_0, z_0) = (5,5,5)$")
plt.savefig("./figures/lorenz_trajectory.pdf")
plt.show()
```

```python
plt.scatter(lorenz_data["trajectory"][:,0], lorenz_data["trajectory"][:,1], s=0.5)
plt.xlabel("$x(t)$")
plt.ylabel("$y(t)$")
plt.title("Lorenz system states $(x(t),y(t)), 1\leq t\leq 51$:\n $(σ, β, ρ) = (10, 8/3, 28), (x_0, y_0, z_0) = (5,5,5)$")
plt.savefig("./figures/lorenz_attractor.pdf")
plt.show()
```

# Bifurcation Diagrams

## Logistic

```python
from LogisticMapLCE import logistic_lce
rr = np.sort(np.random.uniform(3.5, 4, 1000))
logistic_dataset = [logistic_lce(nIterates=1000, includeTrajectory=True, logisticParams=dict(r=r)) for r in rr]
```

```python
for logistic_data in logistic_dataset:
    trajectory = logistic_data["trajectory"]
    r = logistic_data["params"]["r"] * np.ones_like(trajectory)
    plt.scatter(r, trajectory, c="C0", s=0.0001)
plt.title("Logistic map bifurcation diagram $x_n$ vs $r$")
plt.xlabel("Control parameter $r$")
plt.ylabel("State variable $x_n$")
plt.savefig("./figures/logistic_bifurcation.png", dpi=300)
plt.show()
```

## Hénon

```python
from HenonMapLCE import henon_lce
aa = np.sort(np.random.uniform(0.8, 1.4, 1000))
henon_dataset = [henon_lce(nIterates=1000, includeTrajectory=True, henonParams=dict(a=a, b=0.3)) for a in aa]
```

```python
for henon_data in henon_dataset[::4]:
    trajectory = henon_data["trajectory"][:,1]
    a = henon_data["params"]["a"] * np.ones_like(trajectory)
    plt.scatter(a, trajectory, c="C0", s=0.0005)
plt.title("Hénon map bifurcation diagram $y_n$ vs $a$: $b=0.3$")
plt.xlabel("Control parameter $a$")
plt.ylabel("State variable $y_n$")
plt.savefig("./figures/henon_bifurcation.png", dpi=300)
plt.show()
```

## Lorenz

```python
from LorenzODELCE import lorenz_lce
rr = np.sort(np.random.uniform(0, 100, 400))
lorenz_dataset = [lorenz_lce(nIterates=500, includeTrajectory=True, lorenzParams=dict(sigma=10, b=8/3, R=r)) for r in rr]
```

```python
for lorenz_data in lorenz_dataset[::3]:
    trajectory = lorenz_data["trajectory"][:,2]
    r = lorenz_data["params"]["R"] * np.ones_like(trajectory)
    plt.scatter(r, trajectory, c="C0", s=0.001)
plt.title("Lorenz system bifurcation diagram $z(t)$ vs $ρ$: $(\sigma, β)=(10,8/3)$")
plt.xlabel("Control parameter $ρ$")
plt.ylabel("State variable $z(t)$")
plt.savefig("./figures/lorenz_bifurcation.png", dpi=300)
plt.show()
```

# Lyapunov Exponents over Bifurcation Diagrams

## Logistic

```python
from LogisticMapLCE import logistic_lce
rr = np.sort(np.random.uniform(3.5, 4, 1000))
logistic_dataset = [logistic_lce(nIterates=1000, includeTrajectory=True, logisticParams=dict(r=r)) for r in rr]
```

```python
fig, ax1 = plt.subplots()
fig.suptitle("Logistic map $\lambda_{\max}$ vs $r$")

color1 = "gray"
for logistic_data in logistic_dataset:
    trajectory = logistic_data["trajectory"]
    r = logistic_data["params"]["r"] * np.ones_like(trajectory)
    ax1.scatter(r, trajectory, c=color1, s=0.0001, alpha=0.2)
ax1.tick_params(axis='y', labelcolor=color1)
ax1.set_xlabel("Control parameter $r$")
ax1.set_ylabel("State variable $x_n$", c=color1)

ax2 = ax1.twinx()

lyapunovs = [logistic_data["lce"] for logistic_data in logistic_dataset]
rr = [logistic_data["params"]["r"] for logistic_data in logistic_dataset]
color2 = "red"
ax2.plot(rr, lyapunovs, c=color2, lw=0.75)
ax2.axhline(c=color2, linestyle="-.", lw=0.5)
ax2.set_ylabel("Lyapunov exponent $\lambda_{\max}$", color=color2)
ax2.tick_params(axis="y", labelcolor=color2)
plt.savefig("./figures/logistic_bifurcation_with_lyapunovs.png", dpi=300)
plt.show()
```

```python
from HenonMapLCE import henon_lce
aa = np.sort(np.random.uniform(0.8, 1.4, 1000))
henon_dataset = [henon_lce(nIterates=1000, includeTrajectory=True, henonParams=dict(a=a, b=0.3)) for a in aa]
```
